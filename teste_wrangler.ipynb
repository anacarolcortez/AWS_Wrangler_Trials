```python
import awswrangler as wr
import pandas as pd
```

```python
dados = {'name': ["Daenerys Targaryen", "Drogon Targaryen"], 
       'email': ["motherofdragons@dracarys.com", "drogon@dracarys.com"]}
```

```python
df = pd.DataFrame(data=dados)
```

```python
# Salvar dados no Data Lake
wr.s3.to_parquet(
    df=df,
    path="s3://anadatalab/database",
    dataset=True,
    database="learning",
    table="got"
)
```

```python
# Recuperar dados do S3
df = wr.s3.read_parquet("s3://anadatalab/database", dataset=True)
```

```python
# Recuperar dados do Glue pelo Athena
df = wr.athena.read_sql_query("SELECT * FROM got", database="learning")
```
