```python
import boto3
```

## List buckets

### Client

```python
client = boto3.client('s3')
```

```python
response = client.list_buckets()
```

```python
bucket_obj = response['Buckets']
bucket_obj
```

```python
for item in bucket_obj:
    print(f"name: {item['Name']}")
    print(f"created at: {item['CreationDate']}")
```

## Upload files

```python
def upload_files(file_name, bucket, object_name=None, args=None):
    try:
        if object_name is None:
            object_name = file_name
        resp = client.upload_file(file_name, bucket, object_name, ExtraArgs=args)
        print('File uploaded successfully')
    except Exception as error:
        print(f'Error uploading file. Msg: {error}')
```

```python
upload_files('inauguracoes.csv', 'anadatalab')
```

## Download files

### Resource

```python
import re
```

```python
s3 = boto3.resource('s3', region_name='us-east-1') 
```

```python
bucket = s3.Bucket('anadatalab')
```

```python
def download_files(bucket, key):
    file_name = key.split("/")[1]
    if re.findall("^csvfiles/.*csv$", key):
        client.download_file(bucket, key, file_name)
```

```python
for file in bucket.objects.all():
    download_files(file.bucket_name, file.key)
```
